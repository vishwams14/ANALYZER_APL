```python
import pandas as pd

def print_all_unique_values(csv_file):
    try:
        # Read CSV
        df = pd.read_csv(csv_file)

        print("\n===== UNIQUE VALUES FOR EACH COLUMN =====\n")

        # Loop through all columns
        for col in df.columns:
            print(f"\n--- Column: {col} ---")

            # Get unique values (drop NaN)
            unique_values = df[col].dropna().unique()

            # Convert to DataFrame for table format
            table = pd.DataFrame(unique_values, columns=[col])

            # Print table
            print(table.to_string(index=True))

    except Exception as e:
        print("Error:", e)


# ---- Usage ----
csv_file = "APL_2025_LeagueData.csv"  # replace with your CSV file
print_all_unique_values(csv_file)
```

```python
import pandas as pd

def get_columns(csv_file):
    try:
        df = pd.read_csv(csv_file)
        
        print("\nColumns in the CSV file:\n")
        for i, col in enumerate(df.columns, start=1):
            print(f"{i}. {col}")
    
    except Exception as e:
        print("Error:", e)


# ---- Usage ----
csv_file = "APL_2025_LeagueData.csv"  # replace with your file path
get_columns(csv_file)
```
